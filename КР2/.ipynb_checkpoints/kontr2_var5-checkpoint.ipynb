```python
header = ['age', 'workclass','fnlwgt', 'education', 'education-num',
          'marital-status', 'occupation', 'relationship', 
          'race', 'sex', 'capital-gain', 'capital-loss', 
          'hours-per-week', 'native-country', '50K']
```

Используя pandas и датасет, ответьте на вопросы:

1. У какого workclass доход чаще всего превышает 50K (1 балл)?

```python
svod = pd.crosstab(data['workclass'], data['50K'])
svod['dolya'] = svod[' >50K'] / (svod[' <=50K'] + svod[' >50K'])
svod.sort_values('dolya', ascending=True).iloc[-1].name
```

2. Постройте график зависимости доли людей с доходом больше 50K от возраста (1 балл).

```python
svod_age = pd.crosstab(data['age'], data['50K'])
svod_age['dolya'] = svod_age[' >50K'] / (svod_age[' <=50K'] + svod_age[' >50K'])

plt.plot(svod_age.index.values, svod_age['dolya'], color='b')
plt.xlabel('age')
plt.ylabel('% of >50K')
plt.show()
```

## Задание 3  (1 балл)

Преобразуйте категориальные признаки в бинарные.

код отсюда https://github.com/esolovev/ling2019/blob/master/module2/seminar20-21.ipynb

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for column in data.columns:
    if data[column].dtype == 'object' and len(data[column].unique()) == 2:
        print(column)
        data[column] = le.fit_transform(data[column])
```

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
data.head()
```

Разделите выборку на тренировочную и тестовую (80%/20%) с сохранением баланса классов.

```python
from sklearn.model_selection import train_test_split
```

```python
target = data.pop('50K')
```

```python
'50K' in data.columns
```

```python
RANDOM_STATE = 0
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, 
                                                    random_state=RANDOM_STATE, stratify=target)
```

## Задание 4 (3 балла)

Используя обучающую выборку, подберите модель `sklearn.ensemble.RandomForestClassifier`. Использйте стратегию k-fold и перебор параметров (например, поэксперементируйте с числом деревьев и максимальной глубиной дерева) через gridsearchcv.

Если модели слишком долго обучаются, допускается уменьшить количество объектов в обучающей выборке до 5000-10000 с сохранением баланса классов (если качество модели станет хуже, не страшно), изменив долю тренировочной выборки в задании 3.

код отсюда https://github.com/esolovev/ling2019/blob/master/module2/06.05_random_forest.ipynb

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
```

```python
forest = RandomForestClassifier()
parameters = {'n_estimators': [50, 100, 200, 500], 'max_depth':[1, 10, 100]}
```

```python
estimator = GridSearchCV(forest, parameters, cv=5).fit(X_train, y_train)
```

```python
best = estimator.best_estimator_
best.fit(X_train, y_train)
y_pred = best.predict(X_test)
```

Выведите 3 самых важных признака,

```python
feature_importances = best.feature_importances_

importance = {}
for ind in np.argsort(-feature_importances):
    importance[data.columns[ind]] = feature_importances[ind]
    
res = pd.DataFrame({'feature': [*importance], 'weight': [*importance.values()]})
res.sort_values('weight', ascending=False).loc[0:2]
```

а также precision и recall на тестовой выборке.

```python
print(classification_report(y_test, y_pred))
```
